```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt


missing_rule = ["C0_lower","C0_upper","C0_double","C1_lower","C1_upper","C1_double", 
                "C2_lower","C2_upper","C2_double", "C3_lower","C3_upper", "C3_double",
                "C4_lower","C4_upper","C4_double","C5_lower","C5_upper","C5_double",
                "C6_lower","C6_upper","C6_double","C7_lower","C7_upper","C7_double",
]
```

```python
dataname = "california"
missing_type = "BN"
seed = 1
```

```python
# Lower Quantile 

missing_name = missing_rule[0]

processed_data_path_norm = (
    f"../MNAR/datasets/{dataname}/{missing_type}-{missing_name}_seed-{seed}_max-min_norm.pk"
)
processed_data_path = (
    f"../MNAR/datasets/{dataname}/{missing_type}-{missing_name}_seed-{seed}.pk"
)

with open(processed_data_path_norm, "rb") as f:
    norm_lower_observed_values, norm_lower_observed_masks,norm_lower_gt_masks, norm_lower_eval_length = pickle.load(
            f
    )

with open(processed_data_path, "rb") as f:
    lower_observed_values, lower_observed_masks, lower_gt_masks, lower_eval_length = pickle.load(
            f
    )
```

```python
# upper Quantile 

missing_name = missing_rule[1]

processed_data_path_norm = (
    f"../MNAR/datasets/{dataname}/{missing_type}-{missing_name}_seed-{seed}_max-min_norm.pk"
)
processed_data_path = (
    f"../MNAR/datasets/{dataname}/{missing_type}-{missing_name}_seed-{seed}.pk"
)

with open(processed_data_path_norm, "rb") as f:
    norm_upper_observed_values, norm_upper_observed_masks,norm_upper_gt_masks, norm_upper_eval_length = pickle.load(
            f
    )

with open(processed_data_path, "rb") as f:
    upper_observed_values, upper_observed_masks, upper_gt_masks, upper_eval_length = pickle.load(
            f
    )
```

```python
# double Quantile 

missing_name = missing_rule[2]

processed_data_path_norm = (
    f"../MNAR/datasets/{dataname}/{missing_type}-{missing_name}_seed-{seed}_max-min_norm.pk"
)
processed_data_path = (
    f"../MNAR/datasets/{dataname}/{missing_type}-{missing_name}_seed-{seed}.pk"
)

with open(processed_data_path_norm, "rb") as f:
    norm_double_observed_values, norm_double_observed_masks,norm_double_gt_masks, norm_double_eval_length = pickle.load(
            f
    )

with open(processed_data_path, "rb") as f:
    double_observed_values, double_observed_masks, double_gt_masks, double_eval_length = pickle.load(
            f
    )
```

```python
import numpy as np
def apply_mask(observed_values, gt_masks):
    missing_values = observed_values.copy()
    missing_values[gt_masks == 0] = np.nan
    return missing_values
```

```python
missing_value_lower = apply_mask(norm_lower_observed_values, lower_gt_masks)
missing_value_upper = apply_mask(norm_upper_observed_values, upper_gt_masks)
missing_value_double = apply_mask(norm_double_observed_values, double_gt_masks)
```

```python
fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(10, 4*3))
plt.suptitle("California Column 0 ")

# 绘制 observed_values[:, 0] 的直方图
axs[0].hist(norm_upper_observed_values[:, 0], alpha=0.5, bins=20, density=False, label='Complete Data')
axs[0].set_title("Complete Observed Values")
axs[0].legend()

# 绘制 norm_observed_values[:, 0] 和 missing_value[:, 0] 的直方图
axs[1].hist(norm_upper_observed_values[:, 0], alpha=0.5, bins=20, density=False, label='Normalized Data', color='orange')
axs[1].hist(missing_value_upper[:, 0], alpha=0.5, bins=5, density=False, label='Data with missing value', color='green')
axs[1].set_title("Upper quantile missing")
axs[1].legend()


axs[2].hist(norm_lower_observed_values[:, 0], alpha=0.5, bins=20, density=False, label='Normalized Data', color='orange')
axs[2].hist(missing_value_lower[:, 0], alpha=0.5, bins=20, density=False, label='Data with missing value', color='green')
axs[2].set_title("Lower quantile missing")
axs[2].legend()


axs[3].hist(norm_double_observed_values[:, 0], alpha=0.5, bins=20, density=False, label='Normalized Data', color='orange')
axs[3].hist(missing_value_double[:, 0], alpha=0.5, bins=2, density=False, label='Data with missing value', color='green')
axs[3].set_title("Double quantile missing")
axs[3].legend()

plt.show()
```
